### Feature Engineering Process

```python
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()
```

```python
df.info()
```

```python
# Converting date,month,year into integer type
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df.info()
```

```python
# Dropping the feature Date_of_Journey
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.head(1)
```

In the Arrival_Time feature we dont require the date present in that because Date,Month,Year are separately available to us. So we have to handle this.

```python
df['Arrival_Time'] = df['Arrival_Time'].str.split(' ').str[0]
```

```python
df.head()
```

```python
df.isnull().sum()   # Checking missing value
```

As the feature 'Price' is the output, so the null value there will be handled by the model itself. But we have to handle the null values at 'Route' and 'Total_Stops'  

```python
# Now we will separate Arrival Hour and Arrival Minutes
df['Arrival_hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_minute'] = df['Arrival_Time'].str.split(':').str[1]
```

```python
df.head()
```

```python
# Converting Arrival hour and minute into integer type
df['Arrival_hour'] = df['Arrival_hour'].astype(int)
df['Arrival_minute'] = df['Arrival_minute'].astype(int)
```

```python
df.info()
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df.head(1)
```

```python
# Now we will separate Departure Hour and Departure Minutes
df['Dept_hour'] = df['Dep_Time'].str.split(':').str[0]
df['Dept_minute'] = df['Dep_Time'].str.split(':').str[1]
# Converting Departure hour and minute into integer type
df['Dept_hour'] = df['Dept_hour'].astype(int)
df['Dept_minute'] = df['Dept_minute'].astype(int)
```

```python
df.drop('Dep_Time',axis=1,inplace=True)
```
